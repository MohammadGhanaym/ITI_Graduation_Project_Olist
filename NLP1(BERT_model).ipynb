```python
import pandas as pd
from io import StringIO
import re
from transformers import pipeline
from tqdm import tqdm
```

```python
df= pd.read_csv("/content/order_reviews_sentiment (1).csv")
```

```python
# Keep only non-empty comments
df = df[df["review_comment_massage"].notnull()].reset_index(drop=True)
```

```python
# Use a valid Portuguese sentiment model
sentiment_model = pipeline(
    "sentiment-analysis",
    model="pysentimiento/bertweet-pt-sentiment"
)
```

```python
def classify_sentiment(text):
    try:
        result = sentiment_model(text[:512])[0]
        label = result['label']
        # Only two classes:
        if label == "POS":
            return "Positive"
        else:
            # treat NEG (and NEU) as Negative
            return "Negative"
    except Exception as e:
        return "Negative"

tqdm.pandas()
df["sentiment"] = df["review_comment_massage"].progress_apply(classify_sentiment)

print(df[["review_comment_massage", "sentiment"]].head())
df.to_csv("sentiment_results.csv", index=False)
```
